```python
divs = soup.find_all('div',class_='col-sm-18 col-full-xs countdown-item-content')
divs
```

```python
headings = [div.find('h2') for div in divs]
headings
```

```python
movie_names = [heading.find('a').string for heading in headings]
movie_names
```

```python
movie_years = [heading.find('span',class_='subtle start-year').string for heading in headings]
movie_years
```

```python
years = [int(movie[1:-1]) for movie in movie_years]
years
```

```python
scores = [heading.find('span',class_='tMeterScore').string for heading in headings]
score = [int(s.strip('%')) for s in scores]
score
```

```python
critics = [div.find('div',class_='info critics-consensus').text for div in divs]
critics
```

```python
clear_critics = [con[19:] if con.startswith("Critics Consensus: ") else con for con in critics]
clear_critics
```

```python
director = [div.find('div',class_='info director') for div in divs]
clear_director = ["Unknown" if di.find('a') is None else di.find('a').string for di in director ]
clear_director
```

```python
cast = [div.find('div',class_='info cast') for div in divs]
cast[0].find_all('a')[0].string
```

```python
clear = [c.find_all('a') for c in cast]
clear
```

```python
link = [l.string for l in clear[0]]
link
```

```python
casts = ", ".join(link)
casts
```

```python
clean_cast = []
for c in clear:
    cast_names = [link.string for link in c]
    result = ", ".join(cast_names)
    
    clean_cast.append(result)
clean_cast
    
```

```python
ca = [", ".join([link.string for link in c]) for c in clear]
ca
```

```python
import pandas as pd
```

```python
movies_info = pd.DataFrame()
movies_info['Movie Title'] = movie_names
movies_info['Released Year'] = years
movies_info['Rotten Tomato Score'] = score
movies_info['Critic'] = clear_critics
movies_info['Cast'] = ca
movies_info['Director'] = clear_director
```
